```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import pandas as pd

df = pd.read_csv(f'round-2-island-data-bottle/prices_round_2_day_-1.csv', sep=';')
df["SUNLIGHT-1"] = df["SUNLIGHT"].shift(1)
df["SUNLIGHT-2"] = df["SUNLIGHT"].shift(2)
df["SUNLIGHT-3"] = df["SUNLIGHT"].shift(3)
df["HUMIDITY-1"] = df["HUMIDITY"].shift(1)
df["HUMIDITY-2"] = df["HUMIDITY"].shift(2)
df["HUMIDITY-3"] = df["HUMIDITY"].shift(3)
df["ORCHIDS-1"] = df["ORCHIDS"].shift(1)
df["ORCHIDS-2"] = df["ORCHIDS"].shift(2)
df["ORCHIDS-3"] = df["ORCHIDS"].shift(3)
# Define the features and target variables
df = df[['SUNLIGHT-3', 'SUNLIGHT-2', 'SUNLIGHT-1', 'HUMIDITY-3', 'HUMIDITY-2', "HUMIDITY-1", "ORCHIDS-3", "ORCHIDS-2", "ORCHIDS-1", 'ORCHIDS']]
df.dropna(inplace=True)
features = df[['SUNLIGHT-3', 'SUNLIGHT-2', 'SUNLIGHT-1', 'HUMIDITY-3', 'HUMIDITY-2', "HUMIDITY-1", "ORCHIDS-3", "ORCHIDS-2", "ORCHIDS-1"]]
target = df['ORCHIDS']

# Create an instance of XGBRegressor
model = LinearRegression()

# Fit the model
model.fit(features, target)

predictions = model.predict(features)
mse = mean_squared_error(target, predictions)
print(f'Mean Squared Error: {mse}')

```

```python
import matplotlib.pyplot as plt

plt.plot(predictions, label='Predictions')
plt.plot(target, label='Target')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```

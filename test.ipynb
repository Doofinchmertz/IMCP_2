```python
import pandas as pd
```

```python
days = [-1, 0, 1]
for day in days:
    df = pd.read_csv(f'round-2-island-data-bottle/prices_round_2_day_{day}.csv', sep=';')
    for impact_time in range(0, 50):
        print(f'Day: {day}, Impact time: {impact_time}, sunlight_corr: {df["ORCHIDS"].corr(df["SUNLIGHT"].shift(impact_time))}, humidity_corr: {df["ORCHIDS"].corr(df["HUMIDITY"].shift(impact_time))}')

```
